Little Script to clean and format data to show client

```python
import pandas as pd
# Load geocoded data

data = pd.read_csv("../../data/processed/gbh_geocoded_output.csv")

data2 = pd.read_csv("../../data/processed/processed_output.csv")
```

```python
# Concatenate the two DataFrames
merged_data = pd.merge(data, data2, how="left")

# Drop duplicates based on the columns you want to merge
merged_data_unique = merged_data.drop_duplicates()

merged_data_unique = merged_data_unique.dropna()

merged_data_unique.to_csv("../../data/processed/cleaned_output.csv", index=False)


merged_data_unique


```
